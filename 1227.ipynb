```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

sec=pdr.get_data_yahoo('005930.KS', start='2004-01-01')
msft=pdr.get_data_yahoo('MSFT', start='2004-01-01')
```

```python
import matplotlib.pyplot as plt
plt.plot(sec.index, sec.Close, 'b', label='Samasung Electronic')
plt.plot(msft.index, msft.Close, 'r', label='Microsoft')
plt.legend(loc='best')
plt.show()
```

일간 변동률=(오늘 종가-어제종가)/어제종가 X 100

```python
import matplotlib.pyplot as plt
sec_dpc=(sec['Close']-sec['Close'].shift(1))/sec['Close'].shift(1)*100
sec_dpc.iloc[0]=0 #shift함수를 통해 값을 강제로 끌어내려 처음 값이 없어졌으니 0으로 정해주는 것
plt.hist(sec_dpc, bins=100)
plt.grid(True)
plt.show()
```

```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

sec=pdr.get_data_yahoo('005930.KS', start='2004-01-01')
sec_dpc=(sec['Close']-sec['Close'].shift(1))/sec['Close'].shift(1)*100
sec_dpc.iloc[0]=0
sec_dpc_cp=((100+sec_dpc)/100).cumprod()*100-100

msft=pdr.get_data_yahoo('MSFT', start='2004-01-01')
msft_dpc=(msft['Close']-msft['Close'].shift(1))/msft['Close'].shift(1)*100
msft_dpc.iloc[0]=0
msft_dpc_cp=((100+msft_dpc)/100).cumprod()*100-100
```

```python
import matplotlib.pyplot as plt
plt.plot(sec.index, sec_dpc_cp, 'b', label='Samsung Electronics')
plt.plot(msft.index, msft_dpc_cp, 'r', label='Microsft')
plt.ylabel('Change(%)')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
